```python
df = df.fillna(0) # переводим все значения NaN в 0
```

```python
from collections import defaultdict
import statistics as st
```

```python
d = defaultdict(list)
d
```

```python
df_out = df[['country', '2009']]
df_out
```

```python
for index, row in df_out.iterrows():
  d[row[0]].append(row[1])
```

```python
for i, j in d.items():
  print(f'{i=} - median {st.median(j)} mean = {sum(j)/len(j)}')
```
